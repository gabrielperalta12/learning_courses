# Advanced Python — Zero to Master
## Course index

A deep technical treatment of Python: the object model, metaprogramming, concurrency,
performance, and the CPython runtime.

---

## Learning path

```
THE PYTHON OBJECT MODEL
  01 — Pythonic Code & Data Model      Dunder methods, protocols, operator overloading
  02 — Closures & Decorators           First-class functions, closures, decorator factory pattern
  03 — Iterators & Generators          Iterator protocol, generators, yield from, itertools
  04 — Context Managers                __enter__/__exit__, contextlib, exception handling
  05 — Descriptors & Metaclasses       Descriptor protocol, __get__/__set__, metaclass machinery

CONCURRENCY & PERFORMANCE
  06 — Threading & the GIL            Thread model, GIL, concurrent.futures, thread safety
  07 — Async Programming              asyncio, coroutines, event loop, tasks, queues
  08 — Memory & Performance           Profiling, slots, gc, memory views, numpy interop

PROFESSIONAL PYTHON
  09 — Type System                    Type hints, Protocols, TypeVar, generics, runtime checks
  10 — Testing & Design Patterns      pytest advanced, fixtures, mocking, property-based testing
  11 — Packaging & Tooling            pyproject.toml, uv, build, publish, CI/CD
  12 — CPython Internals              Bytecode, dis, frames, C API, Cython intro
```

## Prerequisites

- Comfortable with Python 3 basics: functions, classes, standard library
- Familiar with OOP concepts
- Some exposure to data structures and algorithms

## Python version

All notebooks target **Python 3.11+**. Features from 3.10+ (structural pattern matching) and 3.12+ (type parameter syntax) are noted explicitly.
