# Causal Experimentation & Causal ML — Course Index

## De la experimentación online a decidir a quién tratar

Este curso conecta tres capas que en producto, growth, pricing y UX se usan en cadena:
primero se corre un **experimento online** para medir el efecto promedio de un cambio;
cuando un experimento limpio no es posible, se recurre a **inferencia causal
observacional** sobre datos ya existentes; y cuando la pregunta deja de ser
"¿funciona en promedio?" y pasa a ser **"¿a quién le funciona, y a quién debo tratar?"**,
se usa **causal ML**.

Este curso tiene una orientación aplicada de negocio. La teoría estadística profunda de
varios de estos métodos (potential outcomes, DAGs, matching, IV, RD, DiD, synthetic
control, heterogeneous effects) ya vive en `causal_inference_course/`, y el diseño
experimental clásico (bloques, factoriales, bandits, tests secuenciales, CUPED) vive en
`experimental_design_course/`. Los notebooks de aquí enlazan a esos cursos para la
derivación matemática y se concentran en el caso de uso de negocio, los pitfalls
prácticos y las decisiones que hay que tomar con datos reales.

```
MÓDULO 1 — EXPERIMENTACIÓN ONLINE (A/B testing en producto/growth)
  01 — A/B Testing Fundamentals      Pipeline OCE, métricas ratio, delta method, novelty/primacy
  02 — Randomization Methods         Hashing/bucketing, randomización estratificada, balance checks
  03 — Power Analysis                Tamaño muestral, MDE, CUPED, métricas de revenue sesgadas
  04 — Primary & Secondary Metrics   OEC, jerarquía de métricas, Goodhart's law
  05 — Guardrail Metrics             Latencia/confianza/organizacionales, non-inferiority testing
  06 — Sample Ratio Mismatch         Test chi-cuadrado, causas, árbol de debugging
  07 — Multiple Testing              FWER vs FDR, Bonferroni, Benjamini-Hochberg
  08 — Peeking & Sequential Risk     Inflación de falsos positivos, always-valid inference
  09 — Segmentation & Heterogeneity  Simpson's paradox, multiplicidad de segmentos
  10 — Uplift & Holdouts             Holdouts de marketing, ghost ads/PSA, incrementalidad

MÓDULO 2 — INFERENCIA CAUSAL OBSERVACIONAL (cuando no hay experimento perfecto)
  11 — DiD (caso de negocio)         Rollouts escalonados, Callaway-Sant'Anna, riesgos de TWFE
  12 — Regression Discontinuity      Cutoffs de negocio, manipulación, selección de bandwidth
  13 — Matching (caso de negocio)    Propensity scores, balance, confusión por autoselección
  14 — Fixed Effects / Panel Data    Efectos de entidad y tiempo, within estimator, SE clusterizados
  15 — Event Study                   Efectos dinámicos alrededor de un evento, pre-trends
  16 — Synthetic Control (negocio)   Donor pools, placebo-in-space, synthetic DiD
  17 — Instrumental Variables        2SLS de negocio, instrumentos débiles, Stock-Yogo

MÓDULO 3 — CAUSAL ML (avanzado: ¿a quién tratar para maximizar impacto?)
  18 — Uplift Modeling               4 cuadrantes, single/two-model, uplift curves, Qini
  19 — Heterogeneous Treatment Eff.  CATE, el problema de no tener ground truth, honestidad
  20 — Causal Forests                Honest splitting, GRF, intervalos de confianza para CATE
  21 — Double Machine Learning       Ortogonalidad de Neyman, cross-fitting, R-learner
  22 — Meta-Learners                 S/T/X/R-learner comparados en el mismo dataset simulado
  23 — Policy Learning               Targeting con presupuesto, policy trees, off-policy evaluation
```
