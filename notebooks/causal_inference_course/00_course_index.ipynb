## Course index

## What this course covers

Statistics answers *"what is the association between X and Y?"*
Causal inference answers *"what would happen to Y if we **intervened** on X?"*

The gap between those two questions is the entire subject: a causal effect is a comparison
of **potential outcomes** — what *did* happen vs. what *would have* happened — and one of
the two is always missing. Every method in this course is a strategy for filling in that
missing counterfactual credibly.

```
FOUNDATIONS
  01 — Potential Outcomes            Rubin causal model, ATE/ATT, the fundamental problem
  02 — Randomized Experiments        Fisher exact inference, Neyman estimator, regression adjustment
  03 — Causal Graphs (DAGs)          d-separation, backdoor/frontdoor, do-calculus
  04 — Confounding & Bad Controls    Colliders, M-bias, selection bias, what (not) to adjust for

OBSERVATIONAL DATA: SELECTION ON OBSERVABLES
  05 — Matching & Subclassification  Exact/Mahalanobis matching, overlap, balance checks
  06 — Propensity Scores & IPW       Rosenbaum-Rubin theorem, weighting, doubly robust AIPW

QUASI-EXPERIMENTAL DESIGNS
  07 — Instrumental Variables        2SLS, LATE, weak instruments
  08 — Regression Discontinuity      Sharp/fuzzy RD, local linear estimation, manipulation tests
  09 — Difference-in-Differences     Parallel trends, event studies, TWFE and its pitfalls
  10 — Synthetic Control             Donor pools, pre-treatment fit, placebo inference

MODERN METHODS
  11 — Heterogeneous Effects & ML    Meta-learners, causal forests, double machine learning
  12 — Sensitivity Analysis          Rosenbaum bounds, E-values, placebo and negative controls
```
