# Data Visualization — From Zero to Master
## Course index

A 15-notebook track that takes you from visual perception theory to a production
growth-analytics dashboard. Theory first, libraries second, design always.

---

## Learning path

```
FOUNDATIONS
  00 — Course index                This file
  01 — Visual perception & design  Gestalt, preattentive attributes, chart selection
  02 — Matplotlib architecture     Figure / Axes / Artist, OO vs pyplot, performance
  03 — Chart anatomy & styling      Spines, ticks, grids, rcParams, style sheets

CHART TYPES
  04 — Univariate distributions    Histogram, KDE, box, violin, strip, ECDF
  05 — Bivariate relationships     Scatter, line, bar, area, hexbin
  06 — Multivariate composition    Bubble, parallel coords, radar, heatmap, small multiples

LIBRARIES
  07 — Seaborn                     High-level statistical: relplot, catplot, jointplot, pairplot
  08 — Plotly                      Interactive: express, graph objects, hover, animations

DESIGN & STORYTELLING
  09 — Subplots & layouts          GridSpec, faceting, small multiples, mosaic
  10 — Annotations & storytelling   Callouts, arrows, reference lines, narrative
  11 — Color, typography, brand    Palettes, accessibility, McKinsey/FT style

ADVANCED
  12 — Specialty charts            Waterfall, sankey, treemap, sunburst, funnel, bump
  13 — Dash dashboards             Multi-page layout, callbacks, live data
  14 — Capstone: growth dashboard  End-to-end: data → design → interactive product
```
