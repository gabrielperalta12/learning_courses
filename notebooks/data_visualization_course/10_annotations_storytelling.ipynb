# 10 — Annotations & Storytelling
**Goal:** Turn a chart from a *description* of the data into an *argument*
about the data — callouts, reference lines, headline titles, narrative flow.

```python
import os
os.chdir('/Volumes/SSD_Gabo/proyectos/growth-analytics')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyArrowPatch

np.random.seed(55)
```

## 1. The three jobs of a chart

Knaflic (*Storytelling with Data*) frames every chart as having three jobs:

1. **What** — the data.
2. **So what** — the *insight* (the title carries this).
3. **Now what** — the *action* (annotations push this).

Most charts only do job 1. The difference between a chart and a *story* is
doing jobs 2 and 3.

## 2. Headline titles — the single biggest change

Replace the descriptive title with a finding.

```python
months = pd.date_range('2024-01-01', periods=12, freq='MS')
rev = (np.linspace(100, 152, 12) + np.random.normal(0, 2, 12))

fig, axes = plt.subplots(1, 2, figsize=(13, 4))
axes[0].plot(months, rev, marker='o', color='steelblue')
axes[0].set_title('Revenue (2024)')
axes[0].spines[['top', 'right']].set_visible(False)

axes[1].plot(months, rev, marker='o', color='steelblue', lw=2)
axes[1].set_title('Revenue grew 52% in 2024, led by EMEA',
                  loc='left', weight='bold', fontsize=13, pad=15)
axes[1].text(months[0], rev[0]-7, 'USD millions, Jan–Dec',
             fontsize=9, color='gray')
axes[1].spines[['top', 'right']].set_visible(False)
plt.tight_layout(); plt.show()
```

## 3. Reference lines — give the reader a yardstick

Lines for: target, baseline, average, last period, threshold, forecast horizon.

```python
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(months, rev, marker='o', color='steelblue', lw=2, label='Actual')
ax.axhline(rev.mean(), color='gray', linestyle='--', lw=1, label='Average')
ax.axhline(140, color='crimson', linestyle=':', lw=1.5, label='Target')
ax.fill_between(months, 140, rev.max()+5, color='crimson', alpha=0.05)
ax.legend(frameon=False, loc='lower right')
ax.spines[['top', 'right']].set_visible(False)
ax.set_title('Above target since June — driven by EMEA launch', loc='left', weight='bold')
plt.tight_layout(); plt.show()
```

## 4. `ax.annotate` — callouts with arrows

The annotation API has two coordinate systems you can mix:

- **data coords** (`xy=(x, y)`) — point at a real data location.
- **axes coords** (`xytext=(0.5, 0.95)`) — place text in a relative position.
        Use `textcoords='axes fraction'` to switch.

```python
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(months, rev, marker='o', color='steelblue', lw=2)
ax.spines[['top', 'right']].set_visible(False)

ax.annotate(
    'EMEA launch',
    xy=(months[5], rev[5]), xycoords='data',
    xytext=(months[5], rev[5] + 8), textcoords='data',
    ha='center', fontsize=10, color='crimson', weight='bold',
    arrowprops=dict(arrowstyle='->', color='crimson', lw=1.5),
)
ax.set_title('Single annotation pointing at a real data point', loc='left', weight='bold')
plt.tight_layout(); plt.show()
```

## 5. Highlighting a region

Use `axvspan` / `axhspan` to highlight a time window or value range.

```python
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(months, rev, marker='o', color='steelblue', lw=2)
ax.axvspan(months[5], months[8], color='crimson', alpha=0.12)
ax.text(months[6], rev.max()+2, 'campaign period', color='crimson', fontsize=9)
ax.spines[['top', 'right']].set_visible(False)
ax.set_title('Highlight a campaign window with axvspan', loc='left', weight='bold')
plt.tight_layout(); plt.show()
```

## 6. Direct labels instead of legends

When a chart has only a few series, labeling the line at its end point reads
faster than looking up a legend.

```python
search = rev + 2
social = rev * 0.6
display = rev * 0.4
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(months, search,  color='steelblue', lw=2)
ax.plot(months, social,  color='crimson',  lw=2)
ax.plot(months, display, color='seagreen', lw=2)

for s, name, c in [(search, 'Search', 'steelblue'),
                  (social, 'Social', 'crimson'),
                  (display,'Display','seagreen')]:
    ax.text(months[-1] + pd.Timedelta(days=2), s[-1], name,
            color=c, va='center', fontsize=10, weight='bold')
ax.set_xlim(months[0], months[-1] + pd.Timedelta(days=20))
ax.spines[['top', 'right']].set_visible(False)
ax.set_title('Direct labels — no legend to look up', loc='left', weight='bold')
plt.tight_layout(); plt.show()
```

## 7. Removing the noise — pre-attentive cleanup

Three-step cleanup that turns a default chart into a story chart:

1. **Mute** everything that is not the focus (gray).
2. **Highlight** the one thing the reader should see (color).
3. **Remove** redundant ink (gridlines, ticks, frames).

```python
fig, ax = plt.subplots(figsize=(9, 4))
for s, c, lbl, w in [(search, 'steelblue', 'Search',  3),
                     (social, 'lightgray','Social',  1),
                     (display,'lightgray','Display', 1)]:
    ax.plot(months, s, color=c, lw=w, label=lbl, alpha=0.9)
ax.spines[['top', 'right']].set_visible(False)
ax.grid(axis='y', color='lightgray', alpha=0.5)
ax.legend(frameon=False, loc='lower right')
ax.set_title('Search is 2.5× the next channel — emphasis through de-emphasis',
             loc='left', weight='bold')
plt.tight_layout(); plt.show()
```

## 8. The 6×6 rule and the Big Idea

Knaflic's 6×6: a chart should not need more than 6 seconds to be understood,
and not more than 6 lines of text. Pair every chart with a one-sentence *Big
Idea* that answers the question on the reader's mind.

```python
fig = plt.figure(figsize=(10, 5.5))
gs = GridSpec = __import__('matplotlib.gridspec', fromlist=['GridSpec']).GridSpec
gs_ = GridSpec(2, 1, height_ratios=[0.4, 3.5], hspace=0.05)
ax_text = fig.add_subplot(gs_[0]); ax_text.axis('off')
ax = fig.add_subplot(gs_[1])
ax_text.text(0, 0.8, 'Big idea:', fontsize=11, weight='bold', color='crimson')
ax_text.text(0, 0.2,
             'Q3 revenue beat the plan by 8%, mostly because EMEA paid accounts doubled.',
             fontsize=12, weight='bold')
ax.plot(months, rev, marker='o', color='steelblue', lw=2)
ax.fill_between(months, rev, color='steelblue', alpha=0.2)
ax.axhline(140, color='crimson', linestyle=':', lw=1.2)
ax.text(months[0], 141, 'plan', color='crimson', fontsize=9, va='bottom')
ax.spines[['top', 'right']].set_visible(False)
ax.set_title('Chart supports the headline', loc='left', color='gray', fontsize=10)
plt.tight_layout(); plt.show()
```

## 9. Narrative flow — sequence charts for a story

When a *report* is multiple charts, the order should lead the eye:

1. **Headline** (the so what) — text or KPI tile.
2. **Main chart** — the evidence.
3. **Drivers** — small multiples that decompose the main number.
4. **Action** — callout or CTA box.

```python
fig, axd = plt.subplot_mosaic(
    [
        ['kpi', 'kpi', 'kpi'],
        ['main','main','main'],
        ['d1', 'd2', 'd3'],
        ['cta', 'cta', 'cta'],
    ],
    figsize=(11, 7),
    height_ratios=[0.6, 2.2, 1.5, 0.4],
)
axd['kpi'].text(0.02, 0.5, 'Revenue: $152M  (+52% YoY)',
                fontsize=15, weight='bold', va='center', transform=axd['kpi'].transAxes)
axd['kpi'].axis('off')
axd['main'].plot(months, rev, marker='o', color='steelblue', lw=2)
axd['main'].set_title('Monthly revenue', loc='left')
axd['main'].spines[['top','right']].set_visible(False)
for k in ('d1','d2','d3'):
    axd[k].bar(['A','B','C'], np.random.randint(10, 50, 3),
               color='steelblue', edgecolor='white')
    axd[k].spines[['top','right']].set_visible(False)
    axd[k].set_title({'d1':'By region','d2':'By channel','d3':'By product'}[k])
axd['cta'].text(0.02, 0.5,
                'Action: double EMEA sales capacity by Q4',
                fontsize=11, weight='bold', color='white',
                va='center', transform=axd['cta'].transAxes,
                bbox=dict(facecolor='crimson', pad=10, edgecolor='none'))
axd['cta'].axis('off')
plt.tight_layout(); plt.show()
```

## Summary

| Element | Purpose |
|---|---|
| Headline title | The *so what* (the finding) |
| Reference line | Yardstick (target, average, last period) |
| Annotation | Point at the moment that matters |
| `axvspan` / `axhspan` | Highlight a window or range |
| Direct labels | Skip the legend lookup |
| Mute + highlight | Pre-attentive emphasis |
| Big Idea | One sentence, in the top-left of the panel |
