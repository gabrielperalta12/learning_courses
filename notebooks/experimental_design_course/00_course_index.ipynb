## What this course covers

Experimental design answers the question: **how should we collect data so that our estimates are as precise as possible, given a fixed budget?**

The RCT notebook (statistics course 11) covered how to *analyze* a simple A/B test.
This course covers how to *design* experiments — from classical agriculture to online platforms.

```
CLASSICAL DESIGN
  01 — Design Principles         Fisher's 3 principles, orthogonality, estimability
  02 — Blocking Designs          RCBD, Latin square — remove known nuisance variation
  03 — Factorial Designs (2^k)   Main effects, interactions, effect hierarchy
  04 — Fractional Factorial      2^(k-p), resolution, aliasing — screen many factors cheaply

ADVANCED CLASSICAL
  05 — Response Surface Methods  CCD, Box-Behnken — find the optimum, not just "is there an effect"

SEQUENTIAL AND ADAPTIVE
  06 — Sequential Testing        SPRT, group sequential, alpha spending — stop early safely
  07 — Multi-Armed Bandits       Epsilon-greedy, UCB, Thompson sampling — explore/exploit tradeoff

ONLINE EXPERIMENTS
  08 — Online Experimentation    Network effects, switchback, interleaving, variance reduction
  09 — Optimal Design            D-optimality, information matrix — best design when classical structures don't fit
```

## How designs differ

| Design | Controls for | Estimates | Use when |
|---|---|---|---|
| CRD (A/B test) | Nothing | 1 treatment effect | Homogeneous units |
| RCBD | 1 blocking factor | Treatment + block effects | Known nuisance variable |
| Latin Square | 2 blocking factors | Treatment + 2 block effects | Two nuisance variables |
| 2^k factorial | — | All main effects + interactions | Multiple factors, interactions matter |
| 2^(k-p) fractional | — | Main effects + some 2FI | Screening: many factors, few runs |
| RSM (CCD/BBD) | — | Quadratic surface | Optimization near optimum |
| Adaptive / bandit | — | Best arm over time | Online, welfare maximization matters |
