# Machine Learning — From Zero to Master
## Course index

A 16-notebook track that takes you from "what is machine learning?" to a
production-ready model. The course is built around the canonical texts:

- **An Introduction to Statistical Learning (ISLR)** — James, Witten, Hastie,
  Tibshirani (Springer / Stanford)
- **Hands-On Machine Learning with Scikit-Learn, Keras and TensorFlow** —
  Aurélien Géron (O'Reilly)
- **Deep Learning** — Goodfellow, Bengio, Courville (MIT Press)
- **Pattern Recognition and Machine Learning** — Bishop (Springer / Caltech)
- **Probabilistic Machine Learning** — Murphy (Stanford / U of Edinburgh)
- **Understanding Machine Learning** — Shalev-Shwartz, Ben-David (Cambridge)

Plus rigorous open material from MIT 6.036 / 6.867 / 6.S898, Stanford CS229,
Carnegie Mellon 10-601, and Caltech CS156.

## Learning path

```
FOUNDATIONS
  00 — Course index                  This file
  01 — What is machine learning?     ISLR Ch1, Géron Ch1
  02 — Mathematical primer            Lin alg, probability, optimization
  03 — Statistical learning framework ISLR Ch2 (bias-variance, loss, Bayes)

DATA & LINEAR MODELS
  04 — Data preprocessing             Géron Ch2, ISLR Ch3 prep
  05 — Linear regression              ISLR Ch3 (OLS, diagnostics)
  06 — Classification                 ISLR Ch4 (logistic, LDA, QDA)
  07 — Resampling & validation       ISLR Ch5 (CV, bootstrap)
  08 — Regularization & selection    ISLR Ch6 (ridge, lasso, subset)

NONLINEAR & NONPARAMETRIC
  09 — Tree-based methods            ISLR Ch8 (CART, RF, boosting)
  10 — SVMs and kernel methods       ISLR Ch9, Bishop Ch7
  11 — Unsupervised learning         ISLR Ch10-12 (clustering, PCA, MDS)

DEEP LEARNING
  12 — Neural networks               Géron Ch10, Goodfellow Ch6-7
  13 — Deep learning applications    Géron Ch11-15 (CNN, RNN, Transformer)

FRONTIER
  14 — Advanced topics               Time series, recommenders, causal, RL
  15 — Capstone: end-to-end project  Load → EDA → model → ship
```

## How this course connects to the others

| Track | What you import from it |
|---|---|
| `statistics_course` | Probability, sampling, hypothesis tests |
| `numpy_course` | Vectorization, broadcasting, linear algebra |
| `pandas_course` | DataFrame manipulation, groupby, time series |
| `data_visualization_course` | EDA charts, learning curves, diagnostics |
| `sklearn_course` | API recipes for every model — this course is the theory |
| `eda_course` | Workflow surrounding the model |

If you already know the sklearn API, this course is the *why*. If you already
know the theory, the sklearn course is the *how*.

## Environment

Managed with `uv` (see `pyproject.toml`):

```bash
uv sync                          # install all deps
uv run jupyter lab               # start lab
uv run jupyter nbconvert --execute notebook.ipynb --to notebook  # smoke test
```

Core libraries: **numpy**, **pandas**, **scikit-learn**, **statsmodels**,
**matplotlib**, **seaborn**, **plotly**. Deep learning is taught in PyTorch
(install with `uv add torch` when you reach notebook 12).

### Why this set of references?

- **ISLR** is the gentlest rigorous entry. Every chapter is taught with
  intuition first, then math. R labs in the book; we translate to Python.
- **Géron** is the production-friendly bridge. Each chapter is a working
  project; the second half is a deep learning curriculum.
- **Bishop / Murphy** are the graduate-level reference when you need depth.
- **CS229 / 6.036** are the lectures to watch when a chapter feels dense.
- **Fast.ai** is the practitioner antidote when the math feels too far from
  reality.

If you finish this course and read ISLR cover-to-cover, you will be at the
level of a Stanford graduate student in machine learning.
