**Goal:** Know all the standard ways to generate arrays without typing data by hand.

You rarely build arrays element-by-element in real work. Instead you use factory functions.

## 1. Constant arrays — zeros, ones, full

```python
# Useful as placeholders or initial accumulators
print(np.zeros(5))                   # 1D of zeros
print(np.zeros((3, 4)))              # 2D: 3 rows x 4 cols
print()
print(np.ones((2, 3)))               # all ones
print()
print(np.full((3, 3), fill_value=7)) # any constant
```

```python
# _like variants — match the shape of an existing array
data = np.array([[1, 2, 3], [4, 5, 6]])
print(np.zeros_like(data))
print(np.ones_like(data))
```

## 2. Ranges — arange and linspace

```python
# arange: like Python range() but returns an ndarray
print(np.arange(10))           # 0 to 9
print(np.arange(1, 11))        # 1 to 10
print(np.arange(0, 1, 0.1))    # floats with step
print()

# linspace: N evenly-spaced points between start and stop (inclusive)
# Great for generating axis values in charts
print(np.linspace(0, 1, 5))    # 5 points from 0 to 1
print(np.linspace(0, 100, 11)) # 0, 10, 20, ..., 100
```

```python
# arange vs linspace — key difference
# arange: you control the STEP
# linspace: you control the NUMBER OF POINTS

# Generate dates as day offsets from Jan 1
day_offsets = np.arange(0, 90)   # 0..89 → Q1
print('Days in Q1:', day_offsets.size)
```

## 3. Identity and diagonal matrices

```python
print(np.eye(4))           # identity matrix (1s on diagonal)
print()
print(np.diag([1, 2, 3]))  # matrix with given diagonal values
```

## 4. Random arrays — the most used in practice

```python
rng = np.random.default_rng(seed=42)  # reproducible RNG (modern API)

# Uniform [0, 1)
print(rng.random(5))

# Uniform integers
print(rng.integers(low=0, high=100, size=5))

# Normal (Gaussian) — mean=0, std=1
print(rng.standard_normal(5))

# Normal with custom mean and std
print(rng.normal(loc=500, scale=100, size=5))   # cvr-like data
```

```python
# Simulate a month of funnel sessions (5 channels, 30 days)
rng = np.random.default_rng(42)
sessions = rng.integers(200, 1500, size=(30, 5))  # (days, channels)
print('shape:', sessions.shape)
print(sessions[:3])   # first 3 days
```

```python
# Other handy distributions
rng = np.random.default_rng(42)

# Binomial — conversion events given sessions and CVR
n_sessions = 1000
cvr = 0.05
conversions = rng.binomial(n=n_sessions, p=cvr, size=30)
print('Daily conversions (mean %.1f):' % conversions.mean(), conversions[:10])

# Choice — sample from existing values
channels = np.array(['organic', 'paid', 'email', 'social', 'direct'])
sample = rng.choice(channels, size=10)
print('Sampled channels:', sample)
```

## Summary
| Function | Use |
|---|---|
| `np.zeros(shape)` | Placeholder arrays filled with 0 |
| `np.ones(shape)` | Filled with 1 |
| `np.full(shape, val)` | Filled with any constant |
| `np.arange(start, stop, step)` | Range of values |
| `np.linspace(start, stop, n)` | N evenly-spaced points |
| `np.eye(n)` | Identity matrix |
| `rng.random(n)` | Uniform [0,1) |
| `rng.normal(loc, scale, size)` | Normal distribution |
| `rng.integers(lo, hi, size)` | Random integers |
| `series.to_numpy()` | pandas → NumPy |
