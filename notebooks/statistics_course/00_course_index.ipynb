# Statistics — From Zero to Master
## Course index

This course builds statistical reasoning from the ground up through 12 notebooks,
each with theory, derivations, and reproducible Python code.

---

## Learning path

```
FOUNDATIONS
  01 — Descriptive Statistics    Data types, center, spread, shape, visualization
  02 — Probability Foundations   Axioms, conditional probability, Bayes theorem
  03 — Probability Distributions Discrete + continuous families, CLT preview

CORE INFERENCE
  04 — Sampling and CLT          Sampling distributions, Central Limit Theorem, bootstrap preview
  05 — Estimation                MLE, Method of Moments, confidence intervals, bias-variance
  06 — Hypothesis Testing        t-tests, p-values, Type I/II errors, multiple testing

ADVANCED METHODS
  07 — ANOVA and Non-parametric  One-way ANOVA, Kruskal-Wallis, Mann-Whitney, Wilcoxon
  08 — Regression Analysis       OLS, diagnostics, regularization (Ridge, Lasso)
  09 — Bayesian Statistics       Prior/likelihood/posterior, conjugates, MCMC
  10 — Bootstrap and Resampling  Bootstrap CIs, permutation tests, cross-validation

MASTER LEVEL
  11 — Generalized Linear Models Logistic regression, Poisson regression, GLM framework
  12 — Multivariate Statistics   PCA, biplots, factor analysis, k-means clustering
```

## How this course connects to the Econometrics course

| Statistics concept | Econometrics application |
|---|---|
| OLS (notebook 08) | Gauss-Markov theorem, FWL (notebook 01) |
| CLT + SE (notebook 04) | Inference on beta_hat (notebook 01) |
| Hypothesis testing (notebook 06) | F-test, t-test on coefficients |
| MLE (notebook 05) | IV/2SLS, Logit for selection models |
| Bootstrap (notebook 10) | Wild bootstrap for cluster-robust SE |
| GLMs (notebook 11) | Logit/Probit for binary outcomes |
| Bayesian (notebook 09) | Bayesian IV, shrinkage estimators |
