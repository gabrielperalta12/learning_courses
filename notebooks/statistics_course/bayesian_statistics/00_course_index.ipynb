# Bayesian Statistics — Zero to Master
## Course index

A rigorous and applied treatment of Bayesian statistics following the principal texts
cited by Stanford, CMU, Oxford, and Columbia.

---

## Learning path

```
FOUNDATIONS
  01 — Bayesian Foundations         Philosophy, coherence, prior + likelihood = posterior
  02 — Conjugate Models             Deep dive: all conjugate families, exponential family
  03 — Prior Distributions          Informative, non-informative, Jeffreys, reference, elicitation

COMPUTATION
  04 — Posterior Computation        Laplace approximation, quadrature, importance sampling
  05 — MCMC: Metropolis-Hastings    Theory, random walk MH, independence sampler, convergence diagnostics
  06 — MCMC: Gibbs & HMC           Gibbs sampling, Hamiltonian Monte Carlo, NUTS

MODELING
  07 — Bayesian Model Checking      Posterior predictive checks, Bayes factors, WAIC, LOO-CV
  08 — Hierarchical Models          Partial pooling, hyperpriors, multilevel structure
  09 — Bayesian Regression          Linear regression, variable selection, shrinkage priors

ADVANCED
  10 — Variational Inference        ELBO, mean-field VI, stochastic VI, ADVI
  11 — Gaussian Processes           GP priors, regression, classification, kernel selection
  12 — Bayesian Nonparametrics      Dirichlet process, CRP, DP mixtures, Indian buffet process
```
