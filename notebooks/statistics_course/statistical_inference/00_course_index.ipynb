# Statistical Inference — Zero to Master
## Course index

A rigorous treatment of statistical inference following the canonical graduate texts
cited by MIT, Stanford, CMU, Harvard, and Berkeley.

---

## Learning path

```
PROBABILITY FOUNDATIONS
  01 — Probability Foundations        Axioms, sigma-algebras, conditional probability
  02 — Random Variables               CDFs, PDFs/PMFs, MGFs, characteristic functions
  03 — Multivariate Distributions     Joint, marginal, conditional, independence, copulas
  04 — Convergence & Limit Theorems   LLN, CLT, Slutsky, continuous mapping theorem

ESTIMATION THEORY
  05 — Point Estimation Theory        Sufficiency, completeness, Rao-Blackwell, UMVUE, Cramér-Rao
  06 — Maximum Likelihood             MLE theory, Fisher information, asymptotic normality
  07 — Interval Estimation            Exact CIs, large-sample CIs, likelihood-based intervals

HYPOTHESIS TESTING
  08 — Testing Theory                 NP lemma, UMP tests, power functions, p-values
  09 — Likelihood Ratio Tests         LRT, Wald, Score tests, asymptotic chi-squared

ADVANCED INFERENCE
  10 — Asymptotic Theory              Delta method, asymptotic efficiency, influence functions
  11 — Bayesian Inference             Priors, posteriors, conjugates, Jeffreys, MCMC
  12 — Nonparametric Inference        Bootstrap theory, permutation tests, rank-based tests
  13 — Decision Theory                Loss functions, risk, admissibility, minimax, Bayes rules
```
